```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/home/uday/Uday/Datasets/iris.csv")
df.head()
```

```python
df.shape
```

# Univariate Analysis

```python
df_setosa=df.loc[df["Species"]=="setosa"]
df_virginica=df.loc[df["Species"]=="virginica"]
df_versicolor=df.loc[df["Species"]=="versicolor"]
```

```python
plt.plot(df_setosa["Petal.Length"],np.zeros_like(df_setosa["Petal.Length"]),'o')
plt.plot(df_virginica["Petal.Length"],np.zeros_like(df_virginica["Petal.Length"]),'o')
plt.plot(df_versicolor["Petal.Length"],np.zeros_like(df_versicolor["Petal.Length"]),'o')
plt.xlabel("Petal.Length")
plt.show()
```

# Bivariate Analysis

```python
sns.FacetGrid(df,hue="Species",size=5).map(plt.scatter,"Petal.Length","Sepal.Width").add_legend()
plt.show()
```

# Multivariate Analysis

```python
sns.pairplot(df,hue="Species",size=3)
```
